```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_original = pd.read_csv("/Users/gcohn/Documents/Hybrid-ODE-NeurIPS-2021/results/sample_310_hybrid.txt",skiprows=2, header=None)
df_dcode = pd.read_csv("/Users/gcohn/Documents/Hybrid-ODE-NeurIPS-2021/results/sample_310_hybrid_12_finetuned_5.txt",skiprows=2, header=None)
df = pd.concat([df_original,df_dcode], axis=1)
df.columns=["metric","mean_orig","std_orig","metric","mean_dcode","std_dcode"]
df = df.loc[:,~df.columns.duplicated()].copy()
df = df.set_index('metric')
```

```python
means = df.loc['rmse_x'][['mean_orig', 'mean_dcode']]
std = df.loc['rmse_x'][['std_orig', 'std_dcode']]
plt.ylim(0.63, .67)

plt.xlim(-1, 2)
plt.xticks([])  
plt.title("rmse of Original LHM vs D-CODE + LHM")
plt.ylabel("rmse_x")

plt.errorbar(0, means[0], std[0], fmt='ok', lw=2,color="blue",label="Original LHM")
plt.errorbar(1, means[1], std[1], fmt='ok', lw=2,color="orange",label="D-CODE LHM")

plt.legend(loc="upper left")
```
